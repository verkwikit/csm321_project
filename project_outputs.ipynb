```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
a = 0.9 
b = 0.2 
c = 1.5
d = 0.2 
f = 0.17
Dt = 0.05
```

```python
def intialize():
    global x , xresult , y , yresult ,z , zresult , w , wresult , t , timesteps
    x = 1 
    y = 2 
    z = 0.5
    w =0.5
    
    xresult = [x]
    yresult = [y]
    zresult = [z]
    wresult = [w]
    t = 0
    timesteps = [t]
```

```python
def observe():
    global x , xresult , y , yresult ,z , zresult , w , wresult , t , timesteps
    xresult.append(x)
    yresult.append(y)
    zresult.append(z) 
    wresult.append(w)
    timesteps.append(t)
```

```python
def update():
    global x , xresult , y , yresult ,z , zresult , w , wresult , t , timesteps
    nextx = x + (z + (y-a)*x)*Dt
    nexty = y + (1-b*y-x*x)*Dt
    nextz = z + (-x -c*z)*Dt
    nextw = w + (-d*x*y - f*w)*Dt
    x,y,z,w = nextx,nexty,nextz,nextw
    t = t + Dt
```

```python
intialize()
while t<300.:
    update()
    observe()
    
```

```python
figure()
ax = gca(projection='3d')
ax.plot(xresult,yresult,zresult,'b')
show();
```
